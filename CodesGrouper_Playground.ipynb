# Showcasing CodesGrouper

```python
from ICDCodesGrouper import ICDCodesGrouper
from MimicIII import MimicIII

from config import Settings; settings = Settings()
```

```python
# settings points to the files that contain the mappings: icd9->ccs & icd9->chapters
codes_grouper = ICDCodesGrouper(settings)
```

# Import the Mimic dataset

```python
mimic = MimicIII(settings)

diagnoses = mimic.read_diagnoses().copy()
diagnoses.head(10)
```

# ICD9->CCS

```python
diagnoses['CCS'] = codes_grouper.ccs.lookup(diagnoses['ICD9_CODE'])
diagnoses.head(10)
```

# ICD9->Chapters

```python
diagnoses = mimic.read_diagnoses().copy()
diagnoses.head(10)
```

```python
diagnoses['ICD9 Chapters'] = codes_grouper.icd9chapters.lookup(diagnoses['ICD9_CODE'])
diagnoses.head(10)
```

# Now for all available groups

```python
diagnoses = mimic.read_diagnoses().copy()
diagnoses.head(10)
```

```python
groups = codes_grouper.get_available_groupers()

for g in groups:
    diagnoses[g] = codes_grouper.lookup(g,diagnoses['ICD9_CODE'])
    
diagnoses.head(10)
```
